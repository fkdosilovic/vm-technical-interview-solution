# MNIST: Exploratory data analysis

```python
from miniml import load_mnist
from collections import Counter
from pprint import pprint

from matplotlib import pyplot as plt
```

```python
mnist_path: str = "/home/filip/Data/mnist"
```

```python
train_images, train_labels = load_mnist("/home/filip/Data/mnist", "train")
test_images, test_labels = load_mnist("/home/filip/Data/mnist", "test")
```

```python
print(train_images.shape)
print(test_images.shape)
```

```python
pprint(Counter(train_labels))
pprint(Counter(test_labels))
```

- no class imbalance is present!

```python
print(train_images.min(), train_images.max())
print(test_images.min(), test_images.max())
```

- We need to normalize the inputs. We can do this by dividing the pixel values by 255.

```python
# Plot some digits.
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(train_images[i].reshape(28, 28), cmap="gray")
    ax.set_title(train_labels[i])
```

```python
# Plot normalized digits.
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(train_images[i].reshape(28, 28) / 255, cmap="gray")
    ax.set_title(train_labels[i])
```
